```python
def getWords(filepath):
    with open(filepath) as f:
        lines = f.readlines()
    df = pd.DataFrame(data=lines, columns=['text'])
    df = df.replace(to_replace=r'\/.*$', value='', regex=True)
    df = df.replace(to_replace=r'^\s*\t*', value='', regex=True)
    df = df.replace(to_replace=r"'*", value='', regex=True)
    df = df['text'].str.split(expand=True)
    new = df.values.flatten()
    return np.extract(new != None, new)
```

```python
abh = getWords('./corpora/Abhidharmakosa.txt')
print(abh)
```

```python
def extractPhonemes2(word, i, arr):
    vowel = ['a','ā','e','è','i','o','ö','u','ū']
    terminals = ['ṃ']
    exceptions = ['r', 'ḥ', 'm']
    if len(word) == 0:
        return arr
    if len(word) == i:
        arr.append(word)
        return arr
    if word[i] in vowel:
        #a,e
        if i == 0:
            arr.append(word[:i+1])
            return extractPhonemes2(word[i+1:], 0,arr)
        # next can be a terminal or exception
        else:
            return extractPhonemes2(word, i+1,arr)
    #if 'ṃ'
    if word[i] in terminals:
        arr.append(word[:i+1])
        return extractPhonemes2(word[i+1:], 0,arr)
    #exception is a consonant if first
    if word[i] in exceptions:
        if i == 0:
            return extractPhonemes2(word, i+1, arr)
        else:
            #exceptions are terminals if preceded by a vowel
            if word[i-1] in vowel and i+1 < len(word) and word[i+1] in vowel:
                arr.append(word[:i])
                return extractPhonemes2(word[i:], 0,arr)
            else:
                if i+1 < len(word) and word[i+1] in vowel:
                    return extractPhonemes2(word, i+1, arr)
                else: 
                    arr.append(word[:i+1])
                    return extractPhonemes2(word[i+1:], 0,arr)
    #finally if it is a consonant
    if i == 0:
        return extractPhonemes2(word, i+1, arr)
    else:
        if word[i-1] in vowel:
            arr.append(word[:i])
            return extractPhonemes2(word[i:], 0,arr)
        else:
            return extractPhonemes2(word, i+1, arr)
    return arr
```

```python
all = []
for i in range(len(abh)):
    arr = []
    all = all + extractPhonemes2(abh[i], 0, arr)
print(all)
```
